Q - Find the highest frequency of interval.


```python
test_list = df['mpg']


print ("Original list : " + str(test_list))


max = 0
res = test_list[0]
for i in test_list:
  freq = test_list.count()
  if freq > max:
    max = freq
    res = i

print ("Most frequent number is : " + str(res))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

df = pd.read_csv('mtcars.csv')
kmeans = KMeans(n_clusters=4, random_state=42)
features = df[['mpg', 'wt']]
kmeans.fit(features)
df['cluster'] = kmeans.labels_

outliers = df[(df['wt'] > 5) & (df['mpg'] < 15)]

plt.figure(figsize=(12, 4), dpi=200)

sns.scatterplot(data=df, x='wt', y='mpg', hue='cluster', style='cluster', s=100)
sns.scatterplot(data=outliers, x='wt', y='mpg', marker='X', color='red', s=100)

plt.title("Scatter plot with clusters")
plt.xlabel("Weight of car")
plt.ylabel("Miles per gallon")

plt.show()
```

Q - Plot a boxplot for "price" vs "cut" from the dataset "diamond.csv". Which of the categories under "cut" have the highest median price?


```python
import pandas as pd
import matplotlib.pyplot as plt


diamond_data = pd.read_csv('diamonds.csv')

plt.figure(figsize=(10, 6))
diamond_data.boxplot(column='price', by='cut')
plt.title('Price vs Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()


median_prices = diamond_data.groupby('cut')['price'].median()
highest_median_cut = median_prices.idxmax()
highest_median_price = median_prices.max()

print(f"The cut category '{highest_median_cut}' has the highest median price of {highest_median_price:.2f}.")
```

Q- Show the subplot of the diamond carat weight distribution.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

diamond_data = pd.read_csv('diamonds.csv')


fig, axs = plt.subplots(figsize=(10, 6))


sns.histplot(data=diamond_data, x='carat', kde=True, ax=axs)


axs.set_xlabel('Carat Weight')
axs.set_ylabel('Count')
axs.set_title('Diamond Carat Weight Distribution')


plt.show()
```
